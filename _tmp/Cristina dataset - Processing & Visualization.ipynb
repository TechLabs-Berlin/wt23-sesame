```python
# Add the column "Class" by setting conditional column based on values of the column "Category"

conditions = [
    (df["Category"]== "Eating Out"),
    (df["Category"]=="Shopping"),
    (df["Category"]== "Hobbies"),
    (df["Category"]== "Investment"),
    (df["Category"]== "Travel"),
    (df["Category"]== "Selfcare"),
    (df["Category"]== "Groceries"),
    (df["Category"]== "Rent + Bills"),
    (df["Category"]== "Transport"),
    (df["Category"]== "Study Loan"),
    (df["Category"]== "Healthcare"),
]

values = ["Nice to have", "Nice to have", "Nice to have", "Nice to have", "Nice to have", "Nice to have", "Necessary", "Necessary", "Necessary", "Necessary", "Necessary" ]
df["Class"] = np.select(conditions, values)
print(df)
```

```python
# Add the column "Type"

conditions = [
    (df["Category"]== "Eating Out"),
    (df["Category"]=="Shopping"),
    (df["Category"]== "Hobbies"),
    (df["Category"]== "Investment"),
    (df["Category"]== "Travel"),
    (df["Category"]== "Selfcare"),
    (df["Category"]== "Groceries"),
    (df["Category"]== "Rent + Bills"),
    (df["Category"]== "Transport"),
    (df["Category"]== "Study Loan"),
    (df["Category"]== "Healthcare"),
]

values = ["Variable", "Variable", "Variable", "Variable", "Variable", "Variable", "Variable", "Fixed", "Variable", "Fixed", "Variable"]
df["Type"] = np.select(conditions, values)
print(df)
```

```python
# Change the column "Cost": converting currency with €  to numbers 

series = df["Cost"]
df["Cost"] = pd.to_numeric(series.replace('[^0-9\.-]', "", regex = True))

df
```

```python
# Rename "Cost"

df.rename(columns = {"Cost" : "Cost in €"}, inplace=True)

df
```

```python
# Amount by category

df2 = df.groupby("Category").sum()

# Add the column "% of Total"

df2['% of Total'] = (df2['Cost in €'] / 
                  df2['Cost in €'].sum()) * 100


df2
```
